# MLP Market Decision Example

This is a MLP regression example. This jupyter notebook is based on the tutorial from https://towardsdatascience.com/ml-preface-2-355b1775723e

We are a fictional marketing organization. We look at our historical spend and sales, and map daily marketing spend (this is our X) to daily sales (this is our y). 

In this example, there are two types of marketing spend, brand marketing (brand) to create an image of the brand, and direct to consumer marketing (d2c) to drive sales (i.e., y).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# the dataset is randomly generated
np.random.seed(0)

# randomly generate X
NUM_SAMPLES = int(2e3)
brand_spend = np.random.rand(NUM_SAMPLES)
d2c_spend = np.random.rand(NUM_SAMPLES)

# even the data our X is randomly generated
# but the connections between X and y are following the equations below:
individual_contributions = brand_spend * .01 + d2c_spend * .02
collaboration = np.square(brand_spend) * d2c_spend * 10
sales = individual_contributions + collaboration + np.random.rand(NUM_SAMPLES)/10
```

```python
# let's put X and y together as df
# round number to have 5 digits after decimal point
dataset = pd.DataFrame({
    'brand': brand_spend,
    'd2c': d2c_spend,
    'sales': sales
}).round(5)

dataset.head()
```

```python
# let's see how the data looks like
# with a 3D graph
X = dataset.drop('sales', 1)
Y = dataset['sales']
X1 = X['brand']
X2 = X['d2c']
# dpi Dots Per Inch
fig = plt.figure(dpi=80, figsize = (10, 4))
ax = fig.add_subplot(111, projection='3d')
# s defines the dot sizes in your scatter plot
ax.scatter(X1, X2, Y, c='r', label='real sales', s = 1)
ax.set_xlabel('Brand Marketing')
ax.set_ylabel('Direct to Consumer')
ax.set_zlabel('Sales')
plt.legend()
```

```python
from sklearn.neural_network import MLPRegressor

# let's create a very simple MLP with one hidden layer including 100 hidden neurons 
regr = MLPRegressor(hidden_layer_sizes=(100,))
# this is not very professional here
# because we used all the data for both training and testing
# later we will learn how to fix this issue
regr.fit(X, Y)
Y_hat = regr.predict(X)
```

```python
# let's visually see if our predictions are close to groud truth
fig = plt.figure(dpi=80, figsize=(10, 4))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X1, X2, Y, c='r', label='real sales', s = 1)
ax.scatter(X1, X2, Y_hat, c='b', label='estimated sales', s = 1)
ax.set_xlabel('Brand Marketing')
ax.set_ylabel('Direct to Consumer')
ax.set_zlabel('Sales')
plt.legend()
```

```python
# we can also quantify our errors
from sklearn.metrics import r2_score

# the R2 value (-inf, 1], closer to 1 means better
print(r2_score(Y, Y_hat))
```
