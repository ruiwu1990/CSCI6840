# QQ Plot

This example is based on https://www.yourdatateacher.com/2021/06/21/how-to-use-q-q-plot-for-checking-the-distribution-of-our-data/

It shows how to use a QQ plot to detect if your data follows normal distribution

```python
from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot as plt
# you can customize your figure size and font size with Matplotlib
plt.rcParams['figure.figsize'] = [10, 7]
plt.rc('font', size=14)
from scipy.stats import norm, uniform
import numpy as np
np.random.seed(0)
```

```python
# randomly select 1000 numbers between [1,2)
# this is uniform distribution because each number has
# same probability to be selected
# NOT normal distribution
x = np.random.uniform(1,2,1000)
print(x[:5])
print(len(x))
```

```python
# because x is not normal distribution
# it will be different from normal distribution
# therefore the blue line, i.e., qq plot, is not a 45-degree line
qqplot(x,norm,fit=True,line="45")
plt.show()
```

```python
# if we randomly generate data based on normal distribution
# the qq plot is a 45-degree line
x = np.random.normal(1,2,1000)
qqplot(x,norm,fit=True,line="45")
plt.show()
```
